```python
""" 
Находим интервалы возрастания и убывания.
Первая производная.
f(x)=12·x4·sin(x)·cos(cos(x))-48·x3·sin(cos(x))-54·x2+10·x+10
или
f(x)=2·(6·x4·sin(x)·cos(cos(x))-24·x3·sin(cos(x))-27·x2+5·x+5)
Находим нули функции.
Для этого приравниваем производную к нулю
2·(6·x4·sin(x)·cos(cos(x))-24·x3·sin(cos(x))-27·x2+5·x+5)=0
Откуда:
x1=0.45457357, x2=3.8193108
+------------------------------------------------------------------+
|  (-∞;0.4546)  |0.4546| (0.4546;3.8193)  |3.8193|  (3.8193;+∞)    |
|    f(x)<0	    |f(x)=0|      f(x)>0      |f(x)=0|  f(x)<0         | 
|функция убывает|  min |функция возрастает|  max |  функция убывает|
+------------------------------------------------------------------+
В окрестности точки  x=0.45457357  производная функции меняет знак с (-) на (+).
Следовательно, точка  x=0.45457357  - точка минимума. 
В окрестности точки  x=3.8193108  производная функции меняет знак с (+) на (-). 
Следовательно, точка  x=3.8193108  - точка максимума.
"""

import numpy as np
import matplotlib.pyplot as plt


def f(x):
    return -12 * x**4 * np.sin(np.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30


def f_prime(x):
    return (
        -48 * x**3 * np.sin(np.cos(x)) * np.sin(x)
        - 54 * x**2 * np.cos(x)
        - 54 * x**2 * np.sin(x)
        + 10
    )


def newton_method(x0, tolerance=1e-6, max_iter=10):
    x = x0
    for _ in range(max_iter):
        x = x - f(x) / f_prime(x)
        if abs(f(x)) < tolerance:
            return x
    return None


roots = []
x = np.linspace(-1, 1, 10)
y = np.linspace(-2, 2, 10)
for x0 in x:
    root = newton_method(x0)
    if root is not None:
        roots.append(root)
plt.plot(x, f(x), label="f(x)")
plt.scatter(roots, [0] * len(roots), color="red", label="Корни")
plt.title("График функции")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
